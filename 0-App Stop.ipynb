## Stop Digital Twin Application
This notebook cleanups all Digital Twin resources including app, database instances, pipelines, and tables.

```python
import time
from datetime import datetime

def run_notebook_with_error_display(notebook_path, timeout_seconds=3600):
    """
    Run a notebook and display detailed error information if it fails.
    Returns: (success: bool, result: dict, error_message: str)
    """
    print(f"\n{'='*80}")
    print(f"🚀 Starting: {notebook_path}")
    print(f"{'='*80}")
    
    start_time = time.time()
    
    try:
        result = dbutils.notebook.run(
            notebook_path, 
            timeout_seconds=timeout_seconds
        )
        
        elapsed = time.time() - start_time
        print(f"✅ SUCCESS: {notebook_path}")
        print(f"⏱️  Completed in {elapsed:.2f} seconds")
        print(f"📊 Result: {result}")
        
        return True, result, None
        
    except Exception as e:
        elapsed = time.time() - start_time
        error_msg = str(e)
        
        print(f"\n{'='*80}")
        print(f"❌ FAILED: {notebook_path}")
        print(f"{'='*80}")
        print(f"⏱️  Failed after {elapsed:.2f} seconds")
        print(f"\n📋 Error Details:")
        print(f"{'-'*80}")
        print(error_msg)
        print(f"{'-'*80}")
        
        return False, None, error_msg
```

```python
# Title: Stop - Run Cleanup
from datetime import datetime

print(f"\n🛑 STOPPING DIGITAL TWIN APPLICATION")
print(f"⏰ Started at: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")

print("⚠️  WARNING: This will delete all resources including:")
print("   - Databricks App (digitaltwinapp-sk)")
print("   - Synced Table (latest_sensor_triples)")
print("   - Lakebase Instance (lakebasedigitaltwins-sk)")
print("   - DLT Pipeline (dt_sdp_pipeline)")
print("   - Bronze and Triples tables")
print()

# Confirm before proceeding
response = input("Type 'YES' to confirm cleanup: ")

if response.strip().upper() == "YES":
    print("\n🗑️  Proceeding with cleanup...\n")
    
    success, result, error = run_notebook_with_error_display("./6-Cleanup")
    
    if success:
        print("\n✅ CLEANUP COMPLETED SUCCESSFULLY")
        print("All Digital Twin resources have been removed.")
    else:
        print("\n❌ CLEANUP FAILED")
        print("Some resources may still exist. Please check manually.")
        print(f"\nError: {error}")
else:
    print("\n❌ CLEANUP CANCELLED")
    print("No resources were deleted.")

print(f"\n⏰ Finished at: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
```
